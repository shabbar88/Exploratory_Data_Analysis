# 18.Find Average BasePay of All Employee Per Year

```python
average_basepay_per_year = data.groupby('Year')['BasePay'].mean().reset_index()

print(average_basepay_per_year)
```

# 19. Find Average BasePay of All Employee Per JobTitle

```python
average_basepay_per_jobtitle = data.groupby('JobTitle')['BasePay'].mean().reset_index()

print(average_basepay_per_jobtitle)
```

# 20. Find Average BasePay of Employee Having Job Title ACCOUNTANT

```python
# Filter for employees with the job title ACCOUNTANT
accountants = data[data['JobTitle'] == 'ACCOUNTANT']

# Calculate the average BasePay
average_basepay_accountant = accountants['BasePay'].mean()

print(f"The average BasePay of employees with the job title ACCOUNTANT is: {average_basepay_accountant}")
```
