# 18. Find Top 5 Apps Having Maximum Installs

```python
data['Installs'].dtype
```

```python
data['Installs']
```

```python
data["Installs"].str.replace(",","")
```

```python
# repacing "," with ""
data['Installs1']=data["Installs"].str.replace(",","")
```

```python
# now replacing + with ""
data['Installs1']=data["Installs1"].str.replace("+","")
```

```python
data[data["Installs1"]=="Free"]
```

```python
data["Installs1"]=data["Installs1"].str.replace("Free","0")
```

```python
data['Installs1'].astype(int)
```

```python
# getting top 5 apps having maximum install
data['Installs1'].sort_values(ascending=False).head().index
```

```python
# Store indices of  top 5 Apps Having Maximum Installs
indices=data["Installs1"].sort_values(ascending=False).head().index
```

```python
indices
```

```python
# pass indices (column) to iloc[] or loc[]
data.iloc[indices]["App"] # we get Top 5 App Names having Maximum Installs
```
